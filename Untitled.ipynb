## 用户购买时间和购买

```python
import pandas as pd

data = pd.read_csv('tmall_order_report.csv')
data.head()
```

```python
data.columns = data.columns.str.strip()
data.info()
data1=data.dropna()#订单付款时间为空的代表未购买成功
data1 = data1.copy()
data1['订单创建时间'] = pd.to_datetime(data1['订单创建时间']) 
data1['订单付款时间'] = pd.to_datetime(data1['订单付款时间']) 
```

```python
mask=data1['总金额']!=data1['退款金额']
data2=data1[mask]
data2.head()
#总金额和退款金额相等，可能是买家放弃购买选择退款
```

```python
result = {}
result['总订单数'] = data['订单编号'].count()  
result['已完成订单数'] = data['订单编号'].count()- data1['订单编号'] 
result['未付款订单数'] = data1['订单编号'].count()  
result['退款取消订单数'] = data['订单编号'].count() - data1['订单编号'] -data2['订单编号'] 
result['实际总订单金额'] = data2['买家实际支付金额'].sum()
result['订单创建最早时间']=data2['订单创建时间'].min() 
result['订单付款最晚时间']=data2['订单付款时间'].max() 
```

```python
data2 = data2.copy()  
data2['收货地址'] = data2['收货地址'].str.replace('自治区|维吾尔|回族|壮族|省', '', regex=True)
  # 对省份做个清洗，便于可视化
data2['收货地址'].unique()
```

```python
result2 = data2.groupby('收货地址').agg({'订单编号':'count'}) 
```

```python
from wordcloud import WordCloud  
import numpy as np
import matplotlib.pyplot as plt  


  
# 将 Series 转换为字典，其中键是收货地址，值是订单数量  
word_frequencies = dict(result2['订单编号'])  
# 生成词云图  
wordcloud = WordCloud(font_path="msyh.ttc",width=1000, height=800, background_color='white').generate_from_frequencies(word_frequencies)  
  
# 显示词云图  
plt.figure(figsize=(10, 5))  
plt.imshow(wordcloud, interpolation='bilinear')  
plt.axis('off')  # 不显示坐标轴  
plt.show()
```

```python
data=hourly_transactions['交易量']
from statsmodels.tsa.stattools import adfuller
print('交易量的单位根检验')
result = adfuller(data)#默认情况下，regression参数为'c'，表示使用包含截距项的回归模型。
print('ADF Statistic: %f' % result[0])#ADF统计量
print('p-value: %f' % result[1])#p值
print('Critical Values:')#在置信水平下的临界值
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
print()
```

```python
#差分时序图
diff_data = hourly_transactions['交易量'].diff(periods=1).dropna()# 创建一阶差分的时间序列，加上dropna()后续不需要执行[1:]
#print(diff_data)
 
diff_data.plot()
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot')
#plt.grid()
plt.show()
 

data=diff_data
print('交易量的单位根检验')
result = adfuller(data)#结果对应位置的数据需要自己判断是什么含义
print('ADF Statistic: %f' % result[0])#ADF统计量
print('p-value: %f' % result[1])#p值
print('Critical Values:')#在置信水平下的临界值
for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
print()
 
from statsmodels.stats.diagnostic import acorr_ljungbox
 
#这里为一阶差分后的平稳序列进行白噪声检验，lags为1，否则lags为0，这里拿上述的GDP指标进行
lags = [1,4,8,16,32]
 
print('差分序列的白噪声检验结果为：'+'\n',acorr_ljungbox(diff_data, lags))  # 返回统计量和p值,这里的lags对应差分阶数
 
print('原始数据序列的白噪声检验结果为：'+'\n',acorr_ljungbox(hourly_transactions['交易量'], lags))
```

```python
#生成自相关图，偏自相关图
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
 
fig, ax = plt.subplots(figsize=(5, 4))
plot_acf(diff_data,ax=ax)#可以换成df的数据，这里用一阶差分数据得到平稳序列
    
ax.set_title("Autocorrelation Plot")
lags=list(range(30))
ax.set_xticks(lags)#这里把x坐标刻度变更精细，加网格图更方便,xticklabels替换标签
ax.set_xlabel("Lag")
ax.set_ylabel("Autocorrelation")
ax.grid(alpha=0.5)
plt.legend(['交易量'])#这里区别直接放入df.columns[i]，如果是多字符如‘CS’，这样会被认为是一个序列，拆成C和S的图例
plt.show()
    
fig, ax = plt.subplots(figsize=(5, 4))
lags=list(range(30))
ax.set_xticks(lags)#这里把x坐标刻度变更精细，加网格图更方便,xticklabels替换标签
plot_pacf(diff_data, ax=ax,method='ywm')#ywm替换默认的yw，去除警告
ax.set_xlabel('Lags')
ax.set_ylabel('Partial Autocorrelation')
ax.grid(alpha=0.5)
plt.legend(['交易量'])
plt.show()
 
import warnings
warnings.filterwarnings("ignore")
import statsmodels.api as sm
 
 
#diff_data['GDP'].values.astype(float),这里发现Serise的dtype为object，模型用的应该为float或者int类型，需要注意原数据的数据类型是否一致
 
Min=float('inf')
for i in range(0,6):#AIC，BIC最小找到p，q阶数来定阶，从0开始定阶是否可行？？
    for j in range(0,6):
        result=sm.tsa.ARIMA(hourly_transactions['交易量'].values.astype(float),order=(i,1,j)).fit()
        print([i,j,result.aic,result.bic])
        if result.bic<Min:
            Min=result.bic
            best_pq=[i,j,result.aic,result.bic]
print(f'最优定阶为{best_pq}')
```

```python
Result=sm.tsa.ARIMA(hourly_transactions['交易量'].values.astype(float),order=(best_pq[0],1,best_pq[1])).fit()
print(Result.summary())  #显示模型的所有信息
 
print(len(Result.resid))

lags = [1,4,8,16,32]
print('差分序列的白噪声检验结果为：'+'\n',acorr_ljungbox(Result.resid[1:], lags))
 
## 查看模型的拟合残差分布
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(1,2,1)
plt.plot(Result.resid[1:])
plt.title("ARIMA(4,1,5)残差曲线")
 
## 检查残差是否符合正太分布
ax = fig.add_subplot(1,2,2)
sm.qqplot(Result.resid[1:], line='q', ax=ax)
plt.title("ARIMA(4,1,5)残差Q-Q图")
plt.tight_layout()
plt.show()
 
fig = plt.figure(figsize=(12,5))
Residual=pd.DataFrame(Result.resid[1:])
Residual.plot(kind='kde', title='密度')
plt.legend('')
plt.show()
```

```python
import pandas as pd  
from statsmodels.tsa.arima.model import ARIMA  
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf  
import matplotlib.pyplot as plt  
df=hourly_transactions.copy()
print(df)

# 假设你的数据已经加载到名为df的Da
  
# 检查数据是否平稳（这里只是简单展示，实际分析可能需要更复杂的步骤）  
# 例如，你可以绘制交易量序列的图表，或者计算滚动统计量来观察趋势和季节性  
  
# 如果数据不平稳，你可能需要进行差分或其他转换  
# 这里我们假设数据已经是平稳的，或者我们直接对原始数据进行差分（d=1）  
# 注意：差分应该在建模之前进行，这里只是为了说明  
# df['交易量_diff'] = df['交易量'].diff().dropna()  
  
# 选择ARIMA模型的参数（这里只是示例，实际中需要通过ACF和PACF图等方法来确定）  
p = 4  
d = 1  # 假设我们进行一次差分来使数据平稳  
q = 5
  
# 拟合模型  
model = ARIMA(df['交易量'], order=(p, d, q))  
results = model.fit()  

  
preds = results.get_forecast(steps=24)  
preds_conf_int = preds.conf_int()  
print(preds_conf_int)
```
